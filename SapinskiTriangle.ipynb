```python
from matplotlib import pyplot as plt
import random

coordinates = []
starting_points = [(0,0), (1,2), (2,0)]


coordinates = coordinates + starting_points

def point_on_triangle(pt1, pt2, pt3):
    """
    Random point on the triangle with vertices pt1, pt2 and pt3.
    """
    x, y = sorted([random.random(), random.random()])
    s, t, u = x, y - x, 1 - y
    return (s * pt1[0] + t * pt2[0] + u * pt3[0],
            s * pt1[1] + t * pt2[1] + u * pt3[1])
    


mid_point = point_on_triangle(*starting_points) # Initially Picked at random
coordinates.append(mid_point)

for i in range(2500):
    random_point = random.choice(starting_points)
    mid_point =  ((random_point[0]+mid_point[0])/2, (random_point[1]+mid_point[1])/2)
    coordinates.append(mid_point)

x = [point[0] for point in coordinates]
y = [point[1] for point in coordinates]
```

```python
plt.scatter(x=x, y=y)
plt.show()
```
